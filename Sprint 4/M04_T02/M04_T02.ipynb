Fes les tasques de preprocessat  i adequació del Dataset que disposem en el repositori de GitHub PRE-PROCESSING-DATA amb l'objectiu de preparar-lo i treballar-lo com a dataframe per a extreure'n informació.

```python
#Importamos las librerías
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
# Cargar "movies.dat" en un nou dataframe
df_mov = pd.read_csv('/home/sole/Documents/DS/Sprint 4/movies.dat', sep ='::', encoding = 'ISO-8859-1', engine='python', names = ['Movie_id', 'Title', 'Genre'], index_col='Movie_id')
df_mov.head(10)
```

```python
df_mov.shape
```

```python
df_mov.columns
```

```python
df_mov.Genre.unique()
```

```python
df_mov.Genre.nunique()
```

Podem veure que hi han 301 genres (o combinacions de genres) unics. 

```python
#Analisis estadistic del DF
df_mov.describe()
```

```python
df_mov.info()
```

```python
#Com hi ha pelis amb molt genres, fem que es quedi amb el primer asignat:
df_mov['Genre'] = df_mov.Genre.str.split("|").str[0].str.strip()
df_mov.head(10)
```

```python
df_mov.Genre.nunique()
```

Després d'endreçar els genres, tenim solament 18 genres.

```python
# Ara hem d'extreure l'any de la peli, que per ara apareix en el camp de 'title' i l'hi creem la seva columna: 

df_mov['Year'] =  df_mov['Title'].apply(lambda x: re.findall(r"\((\d{4})\)", x)[0])
df_mov['Title'] = df_mov['Title'].apply(lambda x: re.sub(r"\((\d{4})\)", '', x, 1))
df_mov.head(10)
```

```python
#miro el type de la nova columna
df_mov.info()
```

```python
# Convertim 'Year' de tipus objecte a tipus int per poder utilitzar-la millor en la gestió de les gràfiques
df_mov['Year'] = df_mov.Year.astype(int)
df_mov.info()
```

```python
# En aquest cas tenim 2 variables categòriques i una variable numèrica. Sembla que les representacions gràfiques més il·lustratives seria mostrar la distribució de pel·lícules per cadascun dels gèneres, la de pel·lícules per any i una gràfica on podem veure la distribució dels diferents gèneres en nombre de pel·lícules per cada any.
```

```python
# Pelis per genre en %
plot = (100 * df_mov['Genre'].value_counts() / len(df_mov['Title'])).plot(
    kind='bar', title='Genres per película %')
```

Podem comprovar que gairebé el 60% de les pel·lícules es reparteixen entre el drama i la comèdia, però cal tenir en compte que alguna de les pel·lícules tenia assignats 2 o 3 gèneres i només ens hem quedat amb un, de manera que el resultat de aquest gràfic està condicionat per aquest fet.

```python
#Veiem la distribució de pel·lícules per any en %
plt.figure(figsize=(10,10))
sns.histplot(x='Year', stat="percent",data=df_mov, edgecolor="black")
plt.xticks(rotation=90);
plt.title('Pel·lícules per any %', fontsize=15)
plt.show()
```

Si analitzem en % les pel·lícules realitzades cada any, podem comprovar que és a partir del 1986 quan s'incrementa el nombre de produccions, i l'any 1996 és el més prolífic amb el 17,5% del total de totes les pel·lícules.

```python
plt.figure(figsize=(25,10))
g=sns.boxplot(x='Year', y='Genre', data=df_mov.loc[df_mov.Year >= 1986], color= "lightblue")
g.set_title('Gèneres de pel·lícules per anys a partir de 1986', fontsize=25)
g.set_xlabel('Gèneres',fontsize=20)
g.set_ylabel('Any', fontsize=20)
g.tick_params(labelsize=15)
```

Si posem el focus a les pel·lícules realitzades a partir de 1986, podem veure que hi ha força igualtat entre diversos dels gèneres, podent afirmar que hi ha gèneres que s'han realitzat i es realitzen des dels primers temps del cinema fins a l'actualitat. Altres gèneres tenen temps de vida curts o és possible que responguin a "modes". Per destacar-ne algun, em crida l'atenció el gènere "Musical" que sembla que acaba la vida a partir del 1991, però que va ser majoritari entre el 1987 i el 1989. També el gènere "Film-Noir" desapareix literalment a partir del 1998. En tot cas en aquest gràfic és vàlid el comentari indicat al primer de la biaix que implica triar un dels gèneres per cada pel·lícula.
